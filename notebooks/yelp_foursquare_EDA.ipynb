# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice.Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc) and put your parsed results into a DataFrame

```python
import requests
import pandas as pd

# Foursquare API credentials
foursquare_api_key = "HAITFCA00L2E20WPL0MVETUVXBJEC2ZAVVVCAXUK5GVFNEH"
foursquare_headers = { "accept": "application/json",
    "Authorization": "fsq368Aff4hzILh2OpfOQ42UtsIGhLcmdaj39FpwO7D2Ws0="
    }
foursquare_url = 'https://api.foursquare.com/v3/places/search'

# Coordinates "bixi-toronto"
coordinates = [
    (43.665269, -79.319796),
    (43.671420, -79.445947),
    (43.666224, -79.317693),
    (43.653236, -79.376716),
    (43.663722, -79.380288)
]

def get_foursquare_pois(latitude, longitude, query, limit=50):
    params = {
        'll': f'{latitude},{longitude}',
        'query': query,
        'limit': limit
    }
    response = requests.get(foursquare_url, headers=foursquare_headers, params=params)
    return response.json()

# Lists to hold Foursquare data
foursquare_results = []

# Fetch data from Foursquare for each coordinate
for latitude, longitude in coordinates:
    print(f"Fetching Foursquare data for location: ({latitude}, {longitude})")
    
    for query in ['restaurant', 'bar', 'point of interest']:
        fs_data = get_foursquare_pois(latitude, longitude, query)
        for result in fs_data.get('results', []):
            result['source'] = 'Foursquare'
            result['latitude'] = latitude
            result['longitude'] = longitude
        foursquare_results.extend(fs_data.get('results', []))

# Convert Foursquare results to DataFrame
foursquare_df = pd.json_normalize(foursquare_results)

# Display DataFrame
print("Foursquare DataFrame:")
print(foursquare_df.head())

# Save to CSV file (optional)
foursquare_df.to_csv('foursquare_data.csv', index=False)
```
